```python
df['crime'] = df['crime'].astype(str)
df['crime'] = df['crime'].apply(lambda x: x.zfill(6))
```

```python
df['ShortCrimeCode'] = df['crime'].apply(lambda x: x[:2])
```

```python
### Bring in UCR CSV file here and merge df and it together on 'ShortenedCrime' column
ucr_codes = pd.read_csv('/content/gdrive/My Drive/BoozAllenHamilton/UCRCodes.csv')
```

```python
ucr_codes['UCRCode'] = ucr_codes['UCRCode'].astype(str).apply(lambda x: x.zfill(2))
```

```python
df3 = pd.merge(df, ucr_codes, left_on='ShortCrimeCode', right_on='UCRCode', how = 'left')
df3.drop('UCRCode', axis = 1, inplace=True)
df3
```

```python
# Only pick data points that have a UCRType of 1
df4 = df3.loc[df3['UCRType'] == 1]
df4
```

```python
print(df4.dtypes)
```

```python
# This clears out all missing address, x and y coordinate rows rows
df5 = df4[df4['ileads_address'] != '0.0']
df5 = df5[df5['x_coord'] != 0.0]
df5 = df5[df5['y_coord'] != 0.0]
df5
```

```python
# Fill out all N/A data and turn into -
df['ileads_address'].fillna(0).astype(int)
```

```python
# Turn in all ileads_address points into integers
df5['ileads_address'] = df5['ileads_address'].astype(int)
```

```python
# Grab only points that aren't nan
df5 = df5[df5['ileads_address'] != 'nan']
```

```python
# Took out \.',' from string points of ileads_address
df5['ileads_address'] = df5['ileads_address'].str.replace('\.','')
```

```python
df5['ileads_address'] = df5['ileads_address'].astype(str)
df5
```

```python
df5['address']=df5["ileads_address"]+" "+df5["ileads_street"]+", St.Louis"
df5
```

```python
df5 = df5[df5['ileads_address'] != '0']
df5
```
