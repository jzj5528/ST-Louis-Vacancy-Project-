```python
df = pd.read_csv('/content/gdrive/My Drive/BoozAllenHamilton/CrimeCount(FINAL).csv')
```

```python
df
```

```python
df['Crime Count'] = df['Crime Count'].astype(str)
df
```

```python
df['Crime_Count'] = df['Crime_Count'].astype(float).astype(int)
```

```python
df
```

```python
df.columns = [c.replace(' ', '_') for c in df.columns]
```

```python
df = df[df['Crime_Count'] != 'nan']
df = df[df['Mean_parcelsqft'] != 'nan']
```

```python
df = df[df.Mean_parcelsqft.notnull()]
```

```python
df 
```

```python
df['Percentile_rank']=df.Crime_Count.rank(pct=True)
df
```

```python
df['Quintile']=pd.qcut(df['Crime_Count'], 5, labels=np.arange(1, 6))
df
```

```python
df.to_csv('QuintileCrimeCount.csv', encoding = 'utf-8-sig')
files.download('QuintileCrimeCount.csv')
```
